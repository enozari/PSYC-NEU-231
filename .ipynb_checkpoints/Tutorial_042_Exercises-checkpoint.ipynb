## Read in data from eeg study
* This time I've kept in data across a 1s window from many electrodes (20) sampled at 250Hz, so now we have a [960, 20, 250] matrix that contains data from all 960 trials, 20 electrodes, and over a 1s window following stimulus onset
* Also read in a "cond_labels" vector that has 960 0's and 1's to mark trials from the two experimental conditions 

## import stuff...

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
from sklearn.svm import SVC

fig_font = {'fontname':'Arial', 'fontsize':'20'}
```

```python
# load the data...
eeg = np.load('/Users/Erfan/Documents/GitHub/class_demo_eeg.npz')

# get the different arrays like this...
data = eeg['data']
sr = eeg['sr']
tx = eeg['tx']
cond_labels = eeg['cond_labels']

# shape of data
print('Shape of the big eeg data set: ', data.shape)
print('Number of trials:', len(cond_labels))
print('Number of time points:', len(tx))
```

## Compute the mean across the first dimension (trials) and then plot the mean response across time in each of the 20 electrodes on one axis

* Might need to transpose the data into a Time x Electrode format

```python
data_trialmean = np.mean(data, axis=0)

plt.plot(tx, data_trialmean.T)
plt.show()
```

## That gives you an idea about what the pattern of data looks like across electrodes. 
* Next, take the mean across time, so you have a 960 x 20 matrix 
* Use the cond_labels vector to sort the data based on condition (so you have two 480 x 20 matrices)
* Do classification to see if you can decode the experimental condition
* Remember - break up your data into independent training and testing sets
* Set C to a small number if this is taking a long time...

```python
data_timemean = np.mean(data, axis=2)

d1 = data_timemean[cond_labels==0, :]
d2 = data_timemean[cond_labels==1, :]

# Splitting into train and test
r = d1.shape[0]
n_train = int(np.round(r*0.9))
train_ind = np.random.permutation(r)[0:n_train]
test_ind = np.setdiff1d(np.arange(r), train_ind)
train_data = np.vstack((d1[train_ind, :], d2[train_ind, :]))
test_data = np.vstack((d1[test_ind, :], d2[test_ind, :]))
train_labels = np.hstack((np.zeros(n_train), np.ones(n_train)))
test_labels = np.hstack((np.zeros(r-n_train), np.ones(r-n_train)))

# SVM training
model = SVC(kernel='linear', C=1)
model.fit(train_data, train_labels)

# Testing the model
predict_labels = model.predict(test_data)
predict_acc = np.sum(predict_labels==test_labels) / test_labels.shape[0]
print('SVM prediction accuracy:', predict_acc, '. Barely above chance. Increasing C helps, but too time-consuming for this machine.')
```

## Now try doing this across timepoints!
* First do a time window based on where you think there is nice separation in the responses across electrodes (look at the data that you plotted above)...maybe try 300-400msec post stim
* Then you can try timepoint x timepoint or a moving window

```python
# First plotting the two conditions to see which window seems informative
plt.plot(tx, np.mean(data[cond_labels==0, :, :], axis=0).T, 'b')
plt.plot(tx, np.mean(data[cond_labels==1, :, :], axis=0).T, 'r')
plt.show()

# The suggested 300-400 range seems good1 :) Using that
data_timemean = np.mean(data[:, :, (tx >=300) & (tx <= 400)], axis=2)

d1 = data_timemean[cond_labels==0, :]
d2 = data_timemean[cond_labels==1, :]

# Splitting into train and test
r = d1.shape[0]
n_train = int(np.round(r*0.9))
train_ind = np.random.permutation(r)[0:n_train]
test_ind = np.setdiff1d(np.arange(r), train_ind)
train_data = np.vstack((d1[train_ind, :], d2[train_ind, :]))
test_data = np.vstack((d1[test_ind, :], d2[test_ind, :]))
train_labels = np.hstack((np.zeros(n_train), np.ones(n_train)))
test_labels = np.hstack((np.zeros(r-n_train), np.ones(r-n_train)))

# SVM training
model = SVC(kernel='linear', C=1)
model.fit(train_data, train_labels)

# Testing the model
predict_labels = model.predict(test_data)
predict_acc = np.sum(predict_labels==test_labels) / test_labels.shape[0]
print('SVM prediction accuracy:', predict_acc, '. Clearly better than before!')
```

```python
# Time point by time point is too much for this machine! Trying moving window:
window_len = 100
window_starts = np.arange(0, tx[-1]-window_len, 50)
predict_acc = np.zeros(len(window_starts))

for i in range(len(window_starts)):
    data_timemean = np.mean(data[:, :, (tx >=window_starts[i]) & (tx <= window_starts[i]+window_len)], axis=2)

    d1 = data_timemean[cond_labels==0, :]
    d2 = data_timemean[cond_labels==1, :]

    # Splitting into train and test
    r = d1.shape[0]
    n_train = int(np.round(r*0.9))
    train_ind = np.random.permutation(r)[0:n_train]
    test_ind = np.setdiff1d(np.arange(r), train_ind)
    train_data = np.vstack((d1[train_ind, :], d2[train_ind, :]))
    test_data = np.vstack((d1[test_ind, :], d2[test_ind, :]))
    train_labels = np.hstack((np.zeros(n_train), np.ones(n_train)))
    test_labels = np.hstack((np.zeros(r-n_train), np.ones(r-n_train)))

    # SVM training
    model = SVC(kernel='linear', C=0.01)
    model.fit(train_data, train_labels)

    # Testing the model
    predict_labels = model.predict(test_data)
    predict_acc[i] = np.sum(predict_labels==test_labels) / test_labels.shape[0]

plt.plot(window_starts, predict_acc, '.-', markersize=10)
plt.xlabel('Start time of moving window')
plt.ylabel('Prediction accuracy')
plt.show()

print('We see that 250-350ms and 300-400ms ranges give the best results, as expected from the plot above.')
```

## Bonus - try using randomization testing to eval significance

```python
# Doing it on the 300-400ms range. 
data_timemean = np.mean(data[:, :, (tx >=300) & (tx <= 400)], axis=2)

d1 = data_timemean[cond_labels==0, :]
d2 = data_timemean[cond_labels==1, :]

# Splitting into train and test
r = d1.shape[0]
n_train = int(np.round(r*0.9))
train_ind = np.random.permutation(r)[0:n_train]
test_ind = np.setdiff1d(np.arange(r), train_ind)
train_data = np.vstack((d1[train_ind, :], d2[train_ind, :]))
test_data = np.vstack((d1[test_ind, :], d2[test_ind, :]))
train_labels = np.hstack((np.zeros(n_train), np.ones(n_train)))
test_labels = np.hstack((np.zeros(r-n_train), np.ones(r-n_train)))

# SVM training
model = SVC(kernel='linear', C=0.01)
model.fit(train_data, train_labels)

# Testing the model
predict_labels = model.predict(test_data)
predict_acc = np.sum(predict_labels==test_labels) / test_labels.shape[0]

# Randomization
n_rand = 100
predict_acc_rand = np.zeros(n_rand)

for i in range(n_rand):
    train_labels_rand = np.random.permutation(train_labels)
    model.fit(train_data, train_labels_rand)
    predict_labels = model.predict(test_data)
    predict_acc_rand[i] = np.sum(predict_labels==test_labels) / test_labels.shape[0]
    
plt.hist(predict_acc_rand)
plt.xlabel('prediction accuracy')
plt.show()
    
p = np.sum(predict_acc_rand > predict_acc) / n_rand
print('p-value =', p)
```
